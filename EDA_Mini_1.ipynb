```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the dataset

```python
file_path = '/content/Sample - Superstore.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')
```

1. Monthly sales trends

```python
# Convert date to datetime format
df['Order Date'] = pd.to_datetime(df['Order Date'])
df['Month'] = df['Order Date'].dt.to_period('M')

monthly = df.groupby('Month')['Sales'].sum()
plt.figure(figsize=(10, 5))
monthly.plot(marker='o', linestyle='-')
plt.title('Monthly Sales Trends')
plt.xlabel('Month')
plt.ylabel('Sales')
plt.xticks(rotation=45)
plt.grid()
plt.show()
```

2. Product categories which contribute most revenue

```python
category = df.groupby('Category')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=category.values, y=category.index, palette='coolwarm')
plt.title('Sales by Product Category')
plt.xlabel('Total Sales')
plt.ylabel('Product Category')
plt.show()
```

3. Region wise sales distribution

```python
region = df.groupby('Region')['Sales'].sum()
plt.figure(figsize=(10, 5))
sns.barplot(x=region.values, y=region.index, palette='coolwarm')
plt.title('Region-wise Sales Distribution')
plt.xlabel('Total Sales')
plt.ylabel('Region')
plt.show()
```

4. Customer behaviour impacting sales

```python
customer = df.groupby('Segment')['Sales'].sum().sort_values(ascending=False)
plt.figure(figsize=(10, 5))
sns.barplot(x=customer.values, y=customer.index, palette='coolwarm')
plt.title('Customer type Sales Impact')
plt.xlabel('Total Sales')
plt.ylabel('Customer Segments')
plt.show()
```
