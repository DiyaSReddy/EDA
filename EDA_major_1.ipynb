```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Loading the dataset

```python
file_path = '/content/netflix_titles.csv'
df = pd.read_csv(file_path)
```

1. Count of movies vs. TV shows

```python
print(df.columns)
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='type', data=df, palette='coolwarm')
plt.title('Count of Movies vs. TV Shows')
plt.xlabel('Type')
plt.ylabel('Count')
plt.show()
```

2. Yearly distribution of content

```python
plt.figure(figsize=(10,5))
df['release_year'].value_counts().sort_index().plot(kind='bar', color='teal')
plt.title('Yearly Distribution of Netflix Content')
plt.xlabel('Release Year')
plt.ylabel('Count')
plt.show()
```

3. Most common genres

```python
plt.figure(figsize=(10,5))
genre_counts = df['listed_in'].str.split(', ').explode().value_counts().head(5)
sns.barplot(x=genre_counts.values, y=genre_counts.index, palette='coolwarm')
plt.title('Top 10 Most Common Genres')
plt.xlabel('Count')
plt.ylabel('Genre')
plt.show()
```

4. Top actors and directors

```python
plt.figure(figsize=(10,5))
top_directors = df['director'].dropna().str.split(', ').explode().value_counts().head(5)
sns.barplot(x=top_directors.values, y=top_directors.index, palette='coolwarm')
plt.title('Top Directors')
plt.xlabel('Count')
plt.ylabel('Director')
plt.show()

plt.figure(figsize=(10,5))
top_actors = df['cast'].dropna().str.split(', ').explode().value_counts().head(5)
sns.barplot(x=top_actors.values, y=top_actors.index, palette='coolwarm')
plt.title('Top Actors')
plt.xlabel('Count')
plt.ylabel('Actor')
plt.show()
```
